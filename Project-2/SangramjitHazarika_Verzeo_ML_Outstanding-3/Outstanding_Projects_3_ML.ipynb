# Label Encoding

```python
categorical_cols = {'_golden', 'gender', 'created', 'description',
                    'link_color', 'name', 'profileimage', 'sidebar_color',
                    'text', 'tweet_created', 'tweet_location', 'user_timezone'}
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for each_val in categorical_cols:
    le = LabelEncoder()
    df[each_val] = le.fit_transform(df[each_val])
```

```python
df
```

```python
X = df[['index', '_unit_id', '_golden', '_trusted_judgments'
        , 'gender:confidence', 'profile_yn:confidence',
       'created', 'description', 'fav_number', 'link_color', 'name', 'profileimage', 'retweet_count',
        'sidebar_color', 'text', 'tweet_count', 'tweet_created', 'tweet_location', 'user_timezone']].values

Y = df[['gender']].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, random_state=42)
```

# Feature Scaling using StandardScaler

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# Multi-layer Perceptron classifier in Sklearn

```python
from sklearn.neural_network import MLPClassifier
```

```python
nn = MLPClassifier(activation = "logistic", solver="adam", random_state=42)
```

```python
nn.fit(X_train, Y_train)
```

```python
Y_pred = nn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test, Y_pred)
```
