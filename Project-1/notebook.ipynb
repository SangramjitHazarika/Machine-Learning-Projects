Q3) What is the most common Genre for Vin Diesel & Emma Watson movies?

```python
dict1={}
i=0
df3 = df[df['cast'].str.contains('Vin Diesel') | df['cast'].str.contains('Emma Watson')]
```

```python
df3 #Movies of Vin Diesel & Emma Watson 
```

```python
genre_list=[]
for i in df3['genres']:
    genre = i.split("|")
    for j in range(len(genre)):
        genre_list.append(genre[j])
dict = {}
count = 0
for i in genre_list:
    dict[i]=count
    
for j in genre_list:
    dict[j]=dict[j]+1
dict
```

```python
maximum = max(dict, key=dict.get)
print("The most common Genre for Vin Diesel & Emma Watson movies is:")
print(maximum, "with maximum count ", dict[maximum])
```
